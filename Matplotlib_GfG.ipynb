```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
```

```python
# We will be plotting two lists containing the X, Y coordinates for the plot.
x = [10,20,39,50]
y = [28,38,40,49]

plt.plot(x,y)

plt.title("simple plot")

plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
plt.plot(x)
plt.plot(y, "or")

plt.plot(list(range(0, 22, 3)))
c = [4,3,6,8,3,20,13,15]
plt.plot(c, label = '6th Rep')

plt.title("simple plot")

plt.ylabel("y-axis")
plt.xlabel("x-axis")


ax = plt.gca() #get current axes command

#get command over the individual
#boundary line of the graph body
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#left boundary line to fixed range
ax.spines['left'].set_bounds(-3,40)
# set the interval by which the axis set the marks
plt.xticks(list(range(-3,10)))

plt.yticks(list(range(-3,20,3)))

#legend denotes that what color signifies what
ax.legend(['1st Rep','2nd REp','3rd Rep','4th Rep'])

# plt.annotate('Temperature V/s Days', xy=(1.o1, -2.15))

plt.show()
```

```python
  
a = [1, 2, 3, 4, 5]
b = [0, 0.6, 0.2, 15, 10, 8, 16, 21]
c = [4, 2, 6, 8, 3, 20, 13, 15]
#use fig whenever u want the output in a new window also
# specify in a new window also specify the 
# window size you want ans to be displayed

fig = plt.figure(figsize =(10,10))

  
# creating multiple plots in a 
# single plot
sub1 = plt.subplot(2,2,1)
sub2 = plt.subplot(2,2,2)
sub3 = plt.subplot(2,2,3)
sub4 = plt.subplot(2,2,4)

# sets how the display subplot
# x axis values advances by 1
# dwithin the specified range

sub1.set_xticks(list(range(0,10,1)))
sub1.set_title('1st Rep')

sub2.plot(b, 'or')
sub2.set_xticks(list(range(0, 10, 2)))
sub2.set_title('2nd Rep')

# can directly pass a list in the plot
# function instead adding the reference
sub3.plot(list(range(0, 22, 3)), 'vg')
sub3.set_title('3rd Rep')

sub4.plot(c, 'Dm')
sub4.set_yticks(list(range(0,24,2)))
sub4.set_title('4th Rep')


sub1.plot(a, 'sb')

plt.show()
```

```python
# Figure class:Figure class is the top-level container that contains one or more axes. It is the overall window or page on which everything is drawn.
```

```python
# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize=(5,4))

#creating a new axes for the figure
ax = fig.add_axes([1,1,1,1]) #range of axes

# ax2 = fig.add_axes([1,0.5,0.5,0.5])
# Adding the data to be plotted
ax.plot([2, 3, 4, 5, 5, 6, 6],
        [5, 7, 1, 3, 4, 6 ,8])

plt.show()
```

```python
# 2: Creating multiple plots
```

```python
 
# Creating a new figure with width = 5 inches
# and height = 4 inches

fig = plt.figure(figsize=(5,4))


# Creating first axes for the figure
ax1 = fig.add_axes([1,1,1,1])
# Creating second axes for the figure
ax2 = fig.add_axes([1,0.5,0.5,0.5])

# Adding the data to be plotted
ax1.plot([2, 3, 4, 5, 5, 6, 6],
         [5, 7, 1, 3, 4, 6 ,8])
ax2.plot([1, 2, 3, 4, 5],
         [2, 3, 4, 5, 6])
 
plt.show()
```

```python
# Axes Class:Axes class is the most basic and flexible unit for creating sub-plots. A given figure may contain many axes, but a given axes can only be present in one figure.
```

```python
# Creating the axes object with argument as
# [left, bottom, width, height]
ax = plt.axes([1, 1, 1, 1])

# plotting 1st dataset to the figure
ax1 = ax.plot([1, 2, 3, 4], [1, 2, 3, 4])

# plotting 2nd dataset to the figure
ax2 = ax.plot([1, 2, 3, 4], [2, 3, 4, 5])
```

```python
x = [3, 1, 3]
y = [3, 2, 1]

# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(5, 4))

ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,y)
ax.set_xlim(1,2)

# ax.set_xticklabels(("one", "two", "three", "four", "five", "six"))
plt.show()
```

```python
fig = plt.figure(figsize =(5, 4))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])

# Adding the data to be plotted
ax1.plot([5, 4, 3, 2, 1], [2, 3, 4, 5, 6])
ax2.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6])
plt.show()
```

```python
# Creating the figure and subplots
# according the argument passed
fig, axes = plt.subplots(1, 2)

# plotting the data in the 1st subplot
axes[0].plot([1,2,3,4],[1,2,3,4])

# plotting the data in the 1st subplot only
axes[0].plot([1,2,3,4],[4,3,2,1])

# plotting the data inthe 2nd subplot
axes[1].plot([1,2,3,4],[1,1,1,1])
```
