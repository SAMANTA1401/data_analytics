```python
x =np.arange(1,11)
y=2*x
x,y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("sample")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot(x,y, color='r', linestyle=':', linewidth=2)
plt.show()
```

```python
y1=3*x
```

```python
# adding two lines in the same plot
```

```python
plt.plot(x,y, color='g', linestyle=':', linewidth=2)
plt.plot(x,y1, color='r', linestyle='-.', linewidth=3)
plt.title("Line plot")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.grid(True)
plt.show()
```

```python
# Adding subplot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='g', linestyle=':', linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y1,color='r', linestyle='-.',linewidth=2)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color='g', linestyle=':', linewidth=2)


plt.subplot(2,1,2)
plt.plot(x,y1,color='r', linestyle='-.',linewidth=2)
plt.title('two lines')
plt.show()
```
