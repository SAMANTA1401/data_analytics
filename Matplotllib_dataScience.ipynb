```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7] #no of weekdays
y=[50,51,52,48,47,49,46] # temperature value
```

```python
#line chart
plt.plot(x,y) #matplotlib.pyplot.plot documentation
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color="green", linewidth=5, linestyle="dotted") 
```

```python
plt.plot(x,y, 'r*--')
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y, color='#008040',marker='+', linestyle='',markersize=20)
```

```python
plt.plot(x,y, 'rD', alpha=0.3) #alpha represent tranperency
```

```python
# to plot 3 graph in one chart 

days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,35,35,37]
avg_t = [45,48,46,40,42,41,56]
```

```python
# to give name x axis and y axis call xlabel and ylabel method
#add ligend properties to identify graph line
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

# plt.legend(loc="best")
plt.legend(loc="lower left", shadow=True, fontsize="small")
plt.grid()
```

```python
#bar chart 
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,134,89,27]
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company)) # x position : company name
xpos
```

```python
plt.bar(company,revenue)
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("us Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit,label="profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("us Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("us Tech Stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.legend()
```

```python
#HISTOGRAM: single dimension array . it plots range of value along x axis and freuency along y axis ie how many value belongs into a perticular range
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95) #bins set the range normal:80-100,prediabetics:100-125, diabetic:125-150
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g', histtype='step')
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['g','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['g','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```
