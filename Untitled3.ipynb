```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3,4,5],[1,2,3,2,1])
plt.title("rndm data")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')

x=[1,2,3]
y=[4,6,8]
x2=[6,8,2]
y2=[0,9,7]

plt.plot(x,y, label="first", linewidth=3)
plt.plot(x2,y2, label="second", linewidth=6) 
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3,4,5],[1,2,3,2,1])
plt.title("rndm data")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.show()
```
